```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import time
import config

client_id = config.client_id
client_secret = config.client_secret
client_credentials_manager = spotipy.oauth2.SpotifyClientCredentials(client_id, client_secret)
spotify = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

## カテゴリ取得

```python
# 日本のプレイリストのカテゴリー
results = spotify.categories(country='jp', limit = 50, offset = 0)
li =results['categories']['items']
df = pd.DataFrame.from_dict(li)
df.to_pickle('data/categories.pkl')
df
```

```python
# カテゴリ、travelのプレイリストを見てみる
playlist_data = spotify.category_playlists('travel', country='jp', limit=50)
playlist_df=pd.DataFrame.from_dict(playlist_data['playlists']['items'])
playlist = playlist_df.loc[:, ["name", "description", "id"]]
external_urls = [e['spotify'] for e in playlist_df["external_urls"]]
# playlist['url']=external_urls
```

```python
# 一個目のplaylistの一曲目の情報
tracks = spotify.playlist_tracks(playlist.loc[0, "id"])["items"]
track_dic = pd.DataFrame.from_dict(tracks).loc[:, ["track"]].loc[0, 'track']['album']
track = {key: value for key, value in track_dic.items()
            if key in {'name', 'id', 'artists'}}
```

## プレイリスト取得

```python
error_categories = []
```

```python
# 全カテゴリについてプレイリストを保存
start = 0
for i in range(start, len(df)):
    category = df.loc[i, "id"]
    try: 
        playlist_data = spotify.category_playlists(category, country='jp', limit=50)
    except :
        print("access error:", category)
        error_categories.append(category)
        continue        
    playlist_df=pd.DataFrame.from_dict(playlist_data['playlists']['items'])
    try:
        playlist = playlist_df.loc[:, ["name", "description", "id"]]
        playlist.to_pickle('data/'+ category +'.pkl')
    except :
        print("nodata error:", category)
        error_categories.append(category)
    #print("\r", str(i).zfill(2), str(int(i/len(df)*100)).zfill(3)+"%", category, end="")
    print("\r", str(i).zfill(2), str(int(i/len(df)*100)).zfill(3)+"%", category)
    time.sleep(0.5)
print("100% complete!")
```
